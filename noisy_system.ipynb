## Dynamical system definitions

Dynamical systems are described by
$\dot{x} = f(x,u)$ where $f(x,u)$ can be linear, nonlinear, or whatever.

A linear dynamical system can always be described as
$$\dot{x} = ax + bu$$

An ideal integrator would be:
$$\dot{x} = u$$

Here's a toy exponential system:
$$\dot{x} = \exp(-x + u) - 1$$

Our input, $u$, can be divided into signal and noise components:

$$u=\bar{u} + \eta$$

With a noisy, zero-mean input, we can analyze our system via the Taylor series expansion:
$$\dot{x} = f(x) + \eta f'(x) + \frac{1}{2}\eta^2f''(x) + \frac{1}{3!}\eta^3f^{(3)}(x) + ...$$

Taking the expected value:
$$E[\dot{x}] = E[f(x)] + E[\eta f'(x)] + E[\frac{1}{2}\eta^2f''(x)] + E[\frac{1}{3!}\eta^3f^{(3)}(x)] + ...$$
$$E[\dot{x}] = f(x) + E[\eta]f'(x) + \frac{1}{2}E[\eta^2]f''(x) + \frac{1}{3!}E[\eta^3]f^{(3)}(x) + ...$$

Because $u$ is a zero-mean random variable,
$$E[\dot{x}] = f(x) + E[\eta^2]\frac{1}{2}f''(x) + \frac{1}{4!}E[\eta^4]f^{(4)}(x) + ...$$

Here  we see that only the even ordered derivatives and moments survive. This is why a linear system is not qualitatively affected by noise. For a linear system, $f''(x) = f^{(3)}(x) = f^{(4)} = \ldots = 0$ 

```python
def dxdt_integrator_with_error(x, u):
    return 0.1*np.sin(4*np.pi*x) + u

def dxdt_integrator(x,u):
    return np.zeros(x.shape) + u

def dxdt_exp(x,u):
    return np.exp(-x+u) - 1

def f(x,u):
    return dxdt_exp(x,u)+x
def fd(x,u):
    return -np.exp(-x+u)+1
def fdd(x,u):
    return np.exp(-x+u)
def fddd(x,u):
    return -np.exp(-x+u)
```

## Simulation parameters

```python
### simulation parameters
T = 4       # total simulation time
dt = 0.001  # simulation timestep
x0s = np.linspace(-1.5,1.5,10)  # initial conditions
x = np.linspace(-1.5,1.5,100)   # space to examine phase

### select dynamical system
dxdt = dxdt_exp

### noise parameters 
max_sigma = 1.                      # max noise level
sigmas = np.linspace(0,max_sigma,9) # noise levels
noise_sources = [gaussian(mu=0, sigma=sigma) for sigma in sigmas] # gaussian
# noise_sources = [uniform(mu=0, sigma=sigma) for sigma in sigmas] # uniform

cgen = CG('red') # color generator for plotting
```

## Analyze dynamical system

Plot the phase portrait of the selected dynamical system

```python
plt.figure('phase', figsize=(8,6))
ax = plt.subplot(111)
ax.plot(x, dxdt(x,0), color='k')
plt.axhline(color='b')
s_pts = find_stable_pts(x, dxdt(x,0))
for pt in s_pts:
    plt.axvline(pt)
plt.xlabel(r'$x$', fontsize=18)
plt.ylabel(r'$\dot{x}$', fontsize=18)
plt.xlim((-1.5,1.5))
plt.ylim((-1.0,3.5))
plt.savefig('phase')
```

Plot the phase portrait with noise

```python
s = 1
noise = 1./np.sqrt(2*np.pi*s**2) * np.exp(-x**2/(2*s**2))
plt.figure('phase_noise_overlay', figsize=(10,8)).clf()
gs = gridspec.GridSpec(2, 2, 
                        width_ratios=[4,1], 
                        height_ratios=[4,1])
ax = plt.subplot(gs[0,0])
ax.plot(x, dxdt(x,0), color='k')
plt.axhline(color='b')
s_pts = find_stable_pts(x, dxdt(x,0))
for pt in s_pts:
    plt.axvline(pt)

plt.ylabel(r'$\dot{x}$', fontsize=18)
plt.xlim((-1.5,1.5))
plt.ylim((-1.0,3.5))
plt.tick_params(axis='x', which='both', labelbottom='off')

ax = plt.subplot(gs[1,0])
ax.plot(x, noise, color='m')
ax.axvline(np.mean(noise*x))
plt.xlim((-1.5,1.5))
plt.ylim(((min(noise), max(noise)*1.05)))
plt.xlabel(r'$x$', fontsize=18)
plt.ylabel("input noise\ndistribution", fontsize=12)
plt.tick_params(axis='y', which='both', left='off', right='off', labelleft='off')

ax = plt.subplot(gs[0,1])
ax.plot(noise, dxdt(x,0), color='m')
ax.axhline(np.mean(noise*dxdt(x,0)))
plt.ylim((-1.0,3.5))
plt.xlim((min(noise),max(noise)*1.05))
plt.xlabel("output noise\ndistribution", fontsize=12)
plt.tick_params(axis='x', which='both', bottom='off', top='off', labelbottom='off')
plt.tick_params(axis='y', which='both', labelleft='off')

plt.subplots_adjust(wspace=.02, hspace=.024)
plt.savefig('phase_noise_overlay')
```

```python
plt.figure('phase_noise', figsize=(10,8)).clf()
for noise_idx, noise_src in enumerate(noise_sources):
    plt.subplot(3,3,noise_idx+1)
    dxdt_wnoise = dxdt(x,0) + \
                  sigmas[noise_idx]**2/2. * (dxdt(x,0)+1)  + \
                  sigmas[noise_idx]**4/24. * (dxdt(x,0)+1) + \
                  sigmas[noise_idx]**6/720. * (dxdt(x,0)+1)
    plt.plot(x, dxdt(x,0), color='k')
    plt.plot(x, dxdt_wnoise, color='r')
    plt.axhline(color='b')
    plt.axvline(color='b')
    locs, labels = plt.xticks()
    plt.xticks(locs, locs, rotation=10)
    s_pts = find_stable_pts(x, dxdt_wnoise)
    for pt in s_pts:
        plt.axvline(pt, linestyle=':', color='r')
    plt.xlim((-1.5,1.5))
    plt.ylim((-1.0,3.5))
    plt.title(r'$\sigma=%1.2f$'%sigmas[noise_idx])
plt.subplot(338)
plt.xlabel(r'$x$', fontsize=18)
plt.subplot(334)
plt.ylabel(r'$\dot{x}$', fontsize=18)
plt.tight_layout()
plt.savefig('phase_noise')
```

## Run simulation

Run the dynamical system. Does the system converge to the expected steady state?

```python
plt.figure('phase_with_ICs')
ax = plt.subplot(111)
ax.plot(x, dxdt(x,0), color='k')
ax.set_color_cycle(cgen.get_color_list(10))
for x0 in x0s:
    plt.plot(x0, dxdt(x0,0), 'o', markersize=10)
plt.axhline(color='b')
s_pts = find_stable_pts(x, dxdt(x,0))
for pt in s_pts:
    plt.axvline(pt)
plt.xlabel(r'$x$', fontsize=18)
plt.ylabel(r'$\dot{x}$', fontsize=18)
plt.savefig('phase_with_ICs')
```

```python
plt.figure('dynamics', figsize=(10,8))
xtick_loc = [float(i) for i in range(5)]
xtick_lab = [str(float(i)) for i in range(5)]
for noise_idx, noise_src in enumerate(noise_sources):
    ax = plt.subplot(3,3,noise_idx+1)
    ax.set_color_cycle(cgen.get_color_list(10))
    for x0 in x0s:
        now = 0
        states = [x0]
        time = [now]
        state = x0
        while now < T:
            now = now + dt
            #state = state + dxdt(state, noise_src.get_sample()) * dt # multiplicative noise
            state = state + dxdt(state, 0) * dt + noise_src.get_sample() * np.sqrt(dt) # additive noise
            time.append(now)
            states.append(state)
        ax.plot(time, states)
    plt.xticks(xtick_loc, xtick_loc)
    plt.xlim(0,T)
    plt.axhline(color='b')
    plt.title(r'$\sigma=%1.2f$'%sigmas[noise_idx])
plt.subplot(338)
plt.xlabel(r'$t$', fontsize=18)
plt.subplot(334)
plt.ylabel(r'$x$', fontsize=18)
plt.tight_layout()
plt.savefig('dynamics')
```

Run simulation many times and average over runs

```python
plt.figure('dynamics', figsize=(10,8))
xtick_loc = [float(i) for i in range(5)]
xtick_lab = [str(float(i)) for i in range(5)]
n = T/dt
numtrials = 100
for noise_idx, noise_src in enumerate(noise_sources):
    ax = plt.subplot(3,3,noise_idx+1)
    ax.set_color_cycle(cgen.get_color_list(10))
    for x0 in x0s:
        states = np.zeros((n+1, numtrials))
        states[0,:] = x0
        time = np.arange(0,T+dt/2,dt)
        assert(len(time)==len(states))
        for i in range(numtrials):
            for t in range(len(time)-1):
                states[t+1,i] = states[t,i] + dxdt(states[t,i], 0) * dt + noise_src.get_sample() * np.sqrt(dt) # additive noise
        ax.plot(time, np.mean(states, axis=1))
    plt.xticks(xtick_loc, xtick_loc)
    plt.xlim(0,T)
    plt.axhline(color='b')
    plt.title(r'$\sigma=%1.2f$'%sigmas[noise_idx])
plt.subplot(338)
plt.xlabel(r'$t$', fontsize=18)
plt.subplot(334)
plt.ylabel(r'$x$', fontsize=18)
plt.tight_layout()
plt.savefig('dynamics')
```
