## Ideal integrator with noisy input

Dynamical systems are described by
$\dot{x} = f(x,u)$ where $f(x,u)$ can be linear, nonlinear, or whatever.

A linear dynamical system can always be described as
$$\dot{x} = ax + bu$$

An ideal integrator would be:
$$\dot{x} = u$$

```python
def dxdt_integrator(x,u):
    return np.zeros(x.shape) + u
```

```python
### simulation parameters
T = 4       # total simulation time
dt = 0.001  # simulation timestep
x0 = 0  # initial condition
x = np.linspace(-1.5,1.5,100)   # space to examine phase

### select dynamical system
dxdt = dxdt_integrator

### noise parameters 
max_sigma = .25                      # max noise level
sigmas = np.linspace(0,max_sigma,9) # noise levels
noise_sources = [gaussian(mu=0, sigma=sigma) for sigma in sigmas] # gaussian

cgen = CG('red') # color generator for plotting
```

```python
plt.figure('phase', figsize=(8,6))
ax = plt.subplot(111)
ax.plot(x, dxdt(x,0), color='k')
plt.axhline(color='b')
s_pts = find_stable_pts(x, dxdt(x,0))
for pt in s_pts:
    plt.axvline(pt)
plt.xlabel(r'$x$', fontsize=18)
plt.ylabel(r'$\dot{x}$', fontsize=18)
plt.xlim((-1.5,1.5))
plt.ylim((-.3,.3))
plt.tight_layout()
```

Run the dynamical system. Does the system converge to the expected steady state?

```python
plt.figure('dynamics', figsize=(10,8))
xtick_loc = [float(i) for i in range(5)]
xtick_lab = [str(float(i)) for i in range(5)]
for noise_idx, noise_src in enumerate(noise_sources):
    ax = plt.subplot(3,3,noise_idx+1)
    ax.set_color_cycle(cgen.get_color_list(10))
    for i in range(10):
        now = 0
        states = [x0]
        time = [now]
        state = np.array([x0])
        while now < T:
            now = now + dt
            state = state + noise_src.get_sample() * np.sqrt(dt)
            time.append(now)
            states.append(state)
        ax.plot(time, states)
    plt.xticks(xtick_loc, xtick_loc)
    plt.xlim(0,T)
    plt.ylim((-2,2))
    plt.axhline(color='b')
    plt.title(r'$\sigma=%1.2f$'%sigmas[noise_idx])
plt.subplot(338)
plt.xlabel(r'$t$', fontsize=18)
plt.subplot(334)
plt.ylabel(r'$x$', fontsize=18)
plt.tight_layout()
```

## Integrator with error

```python
import scipy.signal


class interp_system:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    def sample(self,pts):
        return np.interp(pts, self.x, self.y)
```

```python
### simulation parameters
T = 4       # total simulation time
dt = 0.001  # simulation timestep
x0s = np.linspace(-1.5,1.5,10)  # initial conditions
x = np.linspace(-1.5,1.5,100)   # space to examine phase

### noise parameters 
max_sigma = .25                      # max noise level
sigmas = np.linspace(0,max_sigma,9) # noise levels
noise_sources = [gaussian(mu=0, sigma=sigma) for sigma in sigmas] # gaussian

cgen = CG('red') # color generator for plotting
```

```python
# set up system
sys_sigma = .08

x_pts = np.linspace(-3, 3, 200)
x_train = x_pts[0::2]
x_test = x_pts[1::2]

y_train_raw = np.random.normal(0, sys_sigma, len(x_train))
#y_train = y_train_raw
y_train = scipy.signal.wiener(y_train_raw) # add some correlation

sys = interp_system(x_train, y_train)
rmse_train = np.sqrt(np.mean(y_train**2))

y_test = sys.sample(x_test)
rmse_test = np.sqrt(np.mean(y_test**2))


plt.figure('phase', figsize=(8,6)).clf()
ax = plt.subplot(111)
ax.plot(sys.x, sys.y,'k')
ax.plot(x_test, y_test, 'r')
plt.axhline(color='k')
s_pts = find_stable_pts(x_test, y_test)
for pt in s_pts:
    plt.axvline(pt, color='b', linestyle=':')
ax.set_color_cycle(cgen.get_color_list(10))
for x0 in x0s:
    plt.plot(x0, sys.sample(x0), 'o', markersize=10)
plt.xlim([-2,2])
plt.ylim([-.3,.3])
plt.title('Train RMSE=%0.3f'%rmse_train + ' Test RMSE=%0.3f'%rmse_test)
plt.xlabel(r'$x$', fontsize=18)
plt.ylabel(r'$\dot{x}$', fontsize=18)

plt.tight_layout()
```

Run the dynamical system. Does the system converge to the expected steady state?

```python
plt.figure('dynamics', figsize=(10,8))
xtick_loc = [float(i) for i in range(5)]
xtick_lab = [str(float(i)) for i in range(5)]
for noise_idx, noise_src in enumerate(noise_sources):
    ax = plt.subplot(3,3,noise_idx+1)
    ax.set_color_cycle(cgen.get_color_list(10))
    for x0 in x0s:
        now = 0
        states = [x0]
        time = [now]
        state = x0
        while now < T:
            now = now + dt
            state = state + sys.sample(state) * dt + noise_src.get_sample() * np.sqrt(dt) # additive noise
            time.append(now)
            states.append(state)
        ax.plot(time, states)
    plt.axhline(color='b')
    for pt in s_pts:
        plt.axhline(pt, linestyle=':')
    plt.xticks(xtick_loc, xtick_loc)
    plt.xlim(0,T)
    plt.ylim(-2,2)
    plt.title(r'$\sigma=%1.2f$'%sigmas[noise_idx])
plt.subplot(338)
plt.xlabel(r'$t$', fontsize=18)
plt.subplot(334)
plt.ylabel(r'$x$', fontsize=18)
plt.tight_layout()
```
